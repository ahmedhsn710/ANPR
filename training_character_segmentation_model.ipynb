```python
# preprocess the dataset for accurate recognition of characters

# applying dilation to make the characters more prominent/distinguishable and to connect broken/thin chars
def apply_dilation(image):
    kernel = np.ones((3,5), np.uint8) # rectangular kernel shape
    dilated_image = cv2.dilate(image, kernel, iterations=1)
    return Image.fromarray(dilated_image)

# to separate excessively thick characters
def apply_erosion(image):
    kernel = np.ones((3, 3), np.uint8)
    eroded_image = cv2.erode(image, kernel, iterations=1)
    return Image.fromarray(eroded_image)

# eliminates noise/small artifacts to make the characters more distinct and smooth
def apply_opening(image):
    kernel = np.ones((3, 3), np.uint8)
    opened_image = cv2.morphologyEx(image, cv2.MORPH_OPEN, kernel)
    return Image.fromarray(opened_image)

# to enhance the overall shape if characters are broken and to fill gaps
def apply_closing(image):
    kernel = np.ones((3, 3), np.uint8)
    closed_image = cv2.morphologyEx(image, cv2.MORPH_CLOSE, kernel)
    return Image.fromarray(closed_image)

# emphasizes smaller bright details
def apply_tophat(image):
    kernel = np.ones((3, 3), np.uint8)
    tophat_image = cv2.morphologyEx(image, cv2.MORPH_TOPHAT, kernel)
    return Image.fromarray(tophat_image)

# emphasizes smaller white details
def apply_blackhat(image):
    kernel = np.ones((5,5), np.uint8)
    blackhat_image = cv2.morphologyEx(image, cv2.MORPH_BLACKHAT, kernel)
    return Image.fromarray(blackhat_image)

def apply_hybrid_binarization(img):
    # convert PIL Image to numpy array
    image = np.array(img)

    # resize the input image into 4 times
    resized_image = cv2.resize(image, None, fx=4, fy=4, interpolation=cv2.INTER_CUBIC)

    # prform smoothing operation using Positive Laplacian operator
    laplacian = cv2.Laplacian(resized_image, cv2.CV_64F, ksize=3)
    positive_laplacian = np.maximum(0, laplacian)

    # eliminate the background using Gaussian filter
    gaussian_blur = cv2.GaussianBlur(resized_image, (25, 25), 0)
    background_elimination = cv2.subtract(positive_laplacian.astype(np.uint8), gaussian_blur.astype(np.uint8))


    # apply Adaptive bilateral filter
    bilateral_filter = cv2.bilateralFilter(background_elimination, 9, 75, 75)
    adaptive_bilateral_filter = cv2.bilateralFilter(bilateral_filter, 9, 75, 75)

    # convert the image to grayscale
    gray_image = cv2.cvtColor(adaptive_bilateral_filter, cv2.COLOR_BGR2GRAY)

    # binarize the enhanced grayscale image using Otsu's method
    _, binarized_image = cv2.threshold(gray_image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)


    # apply dilation followed by erosion to enhance the characters
    kernel = np.ones((3, 3), np.uint8)
    dilated_image = cv2.dilate(binarized_image, kernel, iterations=1)
    hybrid_binarized_img = cv2.erode(dilated_image, kernel, iterations=1)

    return Image.fromarray(hybrid_binarized_img)

# Gaussian blur to reduce noise and smooth the image
def apply_gaussian_blur(image):
    blurred_image = cv2.GaussianBlur(image, (5, 5), 0)
    return Image.fromarray(blurred_image)

# median blur to reduce noise while preserving edges
def apply_median_blur(image):
    denoised_image = cv2.medianBlur(image, 5)
    return Image.fromarray(denoised_image)

# bilateral filter for noise reduction while preserving edges and details
def apply_bilateral_filter(image):
    denoised_image = cv2.bilateralFilter(image, 9, 75, 75)
    return Image.fromarray(denoised_image)

# histogram equalization to improve contrast in the image
def apply_histogram_equalization(image):
    # Convert the image to grayscale if it's not already
    if len(image.shape) == 3:
        gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    else:
        gray_image = image

    # Apply histogram equalization
    equalized_image = cv2.equalizeHist(gray_image)

    # Convert back to color if the input was a color image
    if len(image.shape) == 3:
        equalized_image = cv2.cvtColor(equalized_image, cv2.COLOR_GRAY2BGR)

    return Image.fromarray(equalized_image)


# CLAHE (Contrast Limited Adaptive Histogram Equalization) for local contrast enhancement
def apply_clahe(image):
    # Convert the image to grayscale if it's not already
    if len(image.shape) == 3:
        gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    else:
        gray_image = image

    # Apply CLAHE
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    clahe_image = clahe.apply(gray_image)

    # Convert back to color if the input was a color image
    if len(image.shape) == 3:
        clahe_image = cv2.cvtColor(clahe_image, cv2.COLOR_GRAY2BGR)

    return Image.fromarray(clahe_image)


# power-law transformation for gamma correction and brightness adjustment
def apply_power_law_transformation(image, gamma=0.7):
    gamma_corrected_image = np.power(image / 255.0, gamma)
    return Image.fromarray((gamma_corrected_image * 255).astype(np.uint8))

# log transformation to enhance details in dark regions of the image
def apply_log_transformation(image):
    log_transformed_image = np.log1p(image)
    return Image.fromarray((log_transformed_image * 255).astype(np.uint8))



def preprocess_image(image_path):
    image = Image.open(image_path)

    # apply_dilation(image)
    # apply_erosion(image)
    # apply_opening(image)
    # apply_closing(image)

    # apply_blackhat(np.array(image))
    # apply_hybrid_binarization(np.array(image))
    # apply_clahe(np.array(image))
    # apply_tophat(np.array(image))
    # apply_power_law_transformation(np.array(image))
    # apply_log_transformation(np.array(image))
    # apply_histogram_equalization(np.array(image))
    apply_bilateral_filter(np.array(image))




    image.save(image_path)


def visualize_images(original, preprocessed, title_original="Original", title_preprocessed="Preprocessed"):
    fig, axes = plt.subplots(1, 2, figsize=(12, 6))

    axes[0].imshow(original)
    axes[0].set_title(title_original)
    axes[0].axis('off')

    axes[1].imshow(preprocessed)
    axes[1].set_title(title_preprocessed)
    axes[1].axis('off')

    plt.show()

# def preprocess_and_visualize_images(image_path):
#     # load the original image
#     original_image = Image.open(image_path)

#     # apply all preprocessing steps
#     dilated_image = apply_dilation(np.array(original_image))
#     eroded_image = apply_erosion(np.array(original_image))
#     opened_image = apply_opening(np.array(original_image))
#     closed_image = apply_closing(np.array(original_image))
#     tophat_image = apply_tophat(np.array(original_image))
#     blackhat_image = apply_blackhat(np.array(original_image))
#     hybrid_binarized_image = apply_hybrid_binarization(np.array(original_image))
#     gaussian_blur_image = apply_gaussian_blur(np.array(original_image))
#     median_blur_image = apply_median_blur(np.array(original_image))
#     bilateral_filter_image = apply_bilateral_filter(np.array(original_image))
#     hitogram_equalized_image = apply_histogram_equalization(np.array(original_image))
#     clahe_image = apply_clahe(np.array(original_image))
#     power_law_image = apply_power_law_transformation(np.array(original_image))
#     log_transformation_image = apply_log_transformation(np.array(original_image))


#     # Visualize the original and preprocessed images
#     visualize_images(original_image, dilated_image, title_original="Original", title_preprocessed="Dilated")
#     visualize_images(original_image, eroded_image, title_original="Original", title_preprocessed="Eroded")
#     visualize_images(original_image, opened_image, title_original="Original", title_preprocessed="Opened")
#     visualize_images(original_image, closed_image, title_original="Original", title_preprocessed="Closed")
#     visualize_images(original_image, tophat_image, title_original="Original", title_preprocessed="Tophat")
#     visualize_images(original_image, blackhat_image, title_original="Original", title_preprocessed="Blackhat")
#     visualize_images(original_image, hybrid_binarized_image, title_original="Original", title_preprocessed="Hybrid Binarization")
#     visualize_images(original_image, gaussian_blur_image, title_original="Original", title_preprocessed="Gaussian Blur")
#     visualize_images(original_image, median_blur_image, title_original="Original", title_preprocessed="Median Blur")
#     visualize_images(original_image, bilateral_filter_image, title_original="Original", title_preprocessed="Bilateral Filter")
#     visualize_images(original_image, clahe_image, title_original="Original", title_preprocessed="CLAHE")
#     visualize_images(original_image, power_law_image, title_original="Original", title_preprocessed="Power Law Transformation")
#     visualize_images(original_image, log_transformation_image, title_original="Original", title_preprocessed="Log Transformation")

# directory path containing the dataset
dataset_directory = "/content/Number-Platers-and-Charaters-4/train/images/"

# preprocess and visualize a few images
num_images_to_visualize = 3

for filename in os.listdir(dataset_directory):
    if filename.endswith(".jpg") or filename.endswith(".png"):
        image_path = os.path.join(dataset_directory, filename)
        # preprocess_and_visualize_images(image_path)
        preprocess_image(image_path)
```
